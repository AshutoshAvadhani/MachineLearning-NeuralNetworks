```python
import numpy as np




def load_data(path):

    def one_hot(y):
        table = np.zeros((y.shape[0], 10))
        for i in range(y.shape[0]):
            table[i][int(y[i][0])] = 1 
        return table

    def normalize(x): 
        x = x / 255
        return x 
    
       
    data = np.genfromtxt(path,delimiter=',',skip_header=1)
    return normalize(data[:,1:]),one_hot(data[:,:1])
    #print(data.shape)
    #print(data[0].shape)
X_train,Y_Train = load_data('mnist_train.csv')
```
